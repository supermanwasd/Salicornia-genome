```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_kin4237_ks = df_kin4237_ks[df_kin4237_ks["ks_NG86"] > 0]
df_kau3287_ks = df_kau3287_ks[df_kau3287_ks["ks_NG86"] > 0]
df_kau3288_ks = df_kau3288_ks[df_kau3288_ks["ks_NG86"] > 0]
df_kin3565_ks = df_kin3565_ks[df_kin3565_ks["ks_NG86"] > 0]
df_kin4218_ks = df_kin4218_ks[df_kin4218_ks["ks_NG86"] > 0]
df_kau3286_ks = df_kau3286_ks[df_kau3286_ks["ks_NG86"] > 0]
df_kin3564_ks = df_kin3564_ks[df_kin3564_ks["ks_NG86"] > 0]
df_Haloxylon_ammodendron = df_Haloxylon_ammodendron[df_Haloxylon_ammodendron["ks_NG86"] > 0]
df_kau3287_Beta_vulgaris = df_kau3287_Beta_vulgaris[df_kau3287_Beta_vulgaris["ks_NG86"] > 0]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create subplots arranged vertically
fig, axs = plt.subplots(3, 1, figsize=(10, 12), dpi=300, sharex=True)

# First subplot
sns.histplot(df_kin3564_ks["ks_NG86"], bins=350, kde=False, color="skyblue", label="S. europaea (France)-S. europaea (France)", ax=axs[0])
#sns.histplot(df_kin4218_ks["ks_NG86"], bins=350, kde=False, color="blue", label="S. herbacea-S. herbacea", ax=axs[0])
sns.histplot(df_kau3286_ks["ks_NG86"], bins=350, kde=False, color="grey", label="S. bigelovii-S. bigelovii", ax=axs[0])
axs[0].set_ylabel("paralogues")
#axs[0].legend(frameon=False)
axs[0].spines['top'].set_visible(False)
axs[0].spines['right'].set_visible(False)

# Second subplot
sns.histplot(df_kau3287_ks["ks_NG86"], bins=350, kde=False, color="#1f77b4", label="S. europaea (Israel)-S. europaea (Israel)", ax=axs[1])
sns.histplot(df_kau3288_ks["ks_NG86"], bins=350, kde=False, color="#ff7f0e", label="S. persica ssp. iranica-S. persica ssp. iranica", ax=axs[1])
sns.histplot(df_kin3565_ks["ks_NG86"], bins=350, kde=False, color="red", label="S. brachiata-S. brachiata", ax=axs[1])
sns.histplot(df_kin4237_ks["ks_NG86"], bins=350, kde=False, color="pink", label="S. sinus-persica-S. sinus-persica", ax=axs[1])
axs[1].set_ylabel("paralogues")
#axs[1].legend(frameon=False)
axs[1].spines['top'].set_visible(False)
axs[1].spines['right'].set_visible(False)


# Third subplot
sns.histplot(df_kau3287_Beta_vulgaris["ks_NG86"], bins=350, kde=False, color="#bcbd22", label="S. europaea (Israel) - Beta vulgaris", ax=axs[2])
axs[2].set_xlabel("Ks")
axs[2].set_ylabel("orthologues")
axs[2].spines['top'].set_visible(False)
axs[2].spines['right'].set_visible(False)
axs[2].axvline(x=0.5484, color="#bcbd22", linestyle="--", linewidth=1)
    
# Set shared x-axis limits
plt.xlim([0, 4])

plt.tight_layout()
#plt.savefig("WGDI.pdf", dpi=300, bbox_inches='tight', format='pdf')
#plt.savefig('WGDI.tif', format='tiff', dpi=300, pil_kwargs={"compression": "tiff_lzw"})
plt.show()
```
